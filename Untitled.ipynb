```python
#choisir une photo aleatoire depuis une list en utilisant glob
```

```python
import glob
import random
from IPython.display import Image, display
import face_recognition
list=glob.glob("photo/*.jpg")
x=random.choice(list)
#print(x)
#display(Image(x))
```

```python
image = face_recognition.load_image_file(x)
face_locations = face_recognition.face_locations(image)
print("I found {} face(s) in this photograph.".format(len(face_locations)))
```

```python
#draw face in picture
from PIL import Image, ImageDraw
pil_image = Image.fromarray(image)
d = ImageDraw.Draw(pil_image)

#left top right bottom
#1 0 3 2
d.rectangle(((face_locations[0][1],face_locations[0][0]),(face_locations[0][3], face_locations[0][2])))
display(pil_image)
```
